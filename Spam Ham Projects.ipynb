```python
import re
import nltk
nltk.download('stopwords')
```

```python
from mylibfornlp.nlppreprocess import NLPProcessor
processor = NLPProcessor()
```

```python
corpus = []
for i in range(0, len(messages)):
    review= processor.preprocess_text(messages['message'][i])
    corpus.append(review)
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
```

```python
words=[]
for sent in corpus:
    sent_token=sent_tokenize(sent)
    for sent in sent_token:
        words.append(simple_preprocess(sent))
```

```python
def avg_word2vec(doc):
    # remove out-of-vocabulary words
    #sent = [word for word in doc if word in model.wv.index_to_key]
    #print(sent)
    
    return np.mean([model.wv[word] for word in doc if word in model.wv.index_to_key],axis=0)
                #or [np.zeros(len(model.wv.index_to_key))], axis=0)
```

```python
from tqdm import tqdm
```

```python
#apply for the entire sentences
import numpy as np
X=[]
for i in tqdm(range(len(words))):
    X.append(avg_word2vec(words[i]))
```

```python
len(X)
```

```python
# Loop through the elements in X and check their lengths
for i, element in enumerate(X):
    # Check if the element is a list or an array
    if isinstance(element, (list, np.ndarray)):
        if len(element) != 100:
            print(f"Element {i} does not have length 100. It has length {len(element)}.")
    else:
        print(f"Element {i} is not a list or array. It is of type {type(element)}.")



```

```python
y = messages[list(map(lambda x: len(x)>0 ,corpus))]
y=pd.get_dummies(y['label'])
y=y.iloc[:,0].values
```

```python
# Initialize a list to store indices of elements to remove
indices_to_remove = []

# Loop through X and check each element's length
for i, element in enumerate(X):
    if isinstance(element, (list, np.ndarray)):
        if len(element) != 100:  # If length is not 100
            indices_to_remove.append(i)  # Store the index
    else:
        indices_to_remove.append(i)  # If the element is not a list or array, store the index

# Remove the elements from X and Y based on the identified indices
X_cleaned = [x for i, x in enumerate(X) if i not in indices_to_remove]
Y_cleaned = [y for i, y in enumerate(y) if i not in indices_to_remove]

# Optional: print the removed indices for verification
print(f"Removed elements at indices: {indices_to_remove}")




```
